```python
#question 5
#create a function to grab the email of the user
def getDomain(email):
  return email.split('@')[-1]
  #returns the last element of string splitted into list
#calling the function
getDomain("user@domain.com")
```

```python
#question 6
def findDog(checkstr):
  return 'dog' in checkstr.lower().split()
  #returns true if 'dog' is an element of the list of the splitted string
#calling the function
findDog('i hate this dog')
```

```python
#question 7
#count the number of occurences of 'dog'
def countDog(checkstr):
  count=0
  for testword in checkstr.lower().split():
    if testword=='dog':
      count+=1
  return count
#calling the function
countDog('hey there dog what its like in NYC, i hate dog but i can love this dog')
```

```python
#question 8
seq=['soup', 'dog', 'salad', 'cat', 'great']
#using filer() function with lambda expression
#filter out the words that don't start with s
list(filter(lambda key: key[0]=='s', seq))
```

```python
#question 9
def speedingChallan(speed, isBirthday):
  if isBirthday:
    speeding=speed-5
  else:
    speeding=speed
  if speeding<=60:
    print('No Challan')
  elif speeding>60 and speeding<=80:
    print('Small Challan')
  elif speeding>80:
    print('Heavy Challan')
#calling the function
speedingChallan(81, True)
speedingChallan(81, False)
speedingChallan(50, False)
```

```python
#question 10
#concatenate two lists index-wise
list1=['M', 'na', 'i', 'She']
list2=['y', 'me', 's', 'lly']
#can be done using zip function
concatlist=[i+j for i, j in zip(list1, list2)]
print(concatlist)
```

```python
#question 11
#concatenate two lists in the form of cross product
list1=['Hello', 'take']
list2=['Dear', 'Sir']
concatlist=[]
for i in list1:
  for j in list2:
    concatlist.append(i+j)
print(concatlist)
```
