```python
#question 8
#creating the matrix
mat=np.arange(1, 26).reshape(5, 5)
print(mat)
```

```python
#extract using indexing and slicing
print(mat[2:5, 1:5])
print(mat[0:3, 1])
```

```python
#get sum of all the values in matrix
sumall=np.sum(mat)
print(sumall)
```

```python
#get sum of all the rows in matrix
sumrow=np.sum(mat, axis=1)
print(sumrow)
```

```python
#get sum of all the columns in the matrix
sumcol=np.sum(mat, axis=0)
print(sumcol)
```
